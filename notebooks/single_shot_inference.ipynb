```python
from pydantic import BaseModel
import os
import joblib

class InputFormat(BaseModel):
    uuid: str
    quote_id: str
    sale_flag: int
    property_type: str
    year_built: int
    number_of_bedrooms: int
    number_of_bathrooms: int
    building_value: float
    contents_value: float
    coverage_level: str
    number_of_occupants: int
    has_smoke_alarms: int
    has_burglar_alarm: int
    pets_present: int
    flood_risk_score: float
    fire_risk_score: float
    crime_rate_score: float
    broker_name: str
    policy_term: int
    previous_claims_count: int
    distance_to_fire_station: float
    has_security_cameras: int
    ownership_status: str
    energy_efficiency_rating: str
    pcd: str
    latitude: float
    longitude: float

model_path = os.getenv('Model_Path')
preprocessor_path = os.getenv('Preprocessor_Path')

if not model_path or not preprocessor_path:
    raise ValueError("Model_Path or Preprocessor_Path environment variable is not set")

class SingleShotInference:

    def __init__(self, inputs: InputFormat):
        self.inputs = inputs
        self.model = self.get_model()
        self.preprocessor = self.get_preprocessor()

    def get_model(self):
        return joblib.load(model_path)

    def get_preprocessor(self):
        return joblib.load(preprocessor_path)

    def predict(self):
        # Ensuring the input data is passed in the correct order
        input_data = [getattr(self.inputs, field) for field in self.inputs.model_fields]
        
        # Preprocessing the input data
        processed_data = self.preprocessor.transform([input_data])
        
        # Making prediction
        prediction = self.model.predict(processed_data)
        
        # Returning prediction
        return prediction[0]

if __name__ == "__main__":
    # Dummy input - replace with real values as needed
    input_data = InputFormat(
        uuid="123",
        quote_id="Q456",
        sale_flag=1,
        property_type="house",
        year_built=2000,
        number_of_bedrooms=3,
        number_of_bathrooms=2,
        building_value=250000.0,
        contents_value=50000.0,
        coverage_level="standard",
        number_of_occupants=4,
        has_smoke_alarms=1,
        has_burglar_alarm=0,
        pets_present=1,
        flood_risk_score=0.2,
        fire_risk_score=0.3,
        crime_rate_score=0.5,
        broker_name="AgentX",
        policy_term=12,
        previous_claims_count=0,
        distance_to_fire_station=1.5,
        has_security_cameras=1,
        ownership_status="owned",
        energy_efficiency_rating="A",
        pcd="AB12CD",
        latitude=51.5074,
        longitude=0.1278
    )

    inference = SingleShotInference(inputs=input_data)
    result = inference.predict()
    print("Prediction:", result)
```
